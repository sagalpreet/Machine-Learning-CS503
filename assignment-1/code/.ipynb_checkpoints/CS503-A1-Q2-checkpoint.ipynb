```python
import numpy as np
import matplotlib.pyplot as plt
import random
from math import ceil, floor

np.random.seed(100)
```

## Polynomial Regression in One Dimension [Easy]

```python
# utilities for creating a dataset

### data structure to represent dataset
class Dataset:
    def __init__(self, x: np.ndarray, y: np.ndarray, n = None):
        # todo: assertion to verify the dimension of x and y
        self.__x = x
        self.__y = y
        self.__n = n if n else len(x)
    
    @property
    def x(self):
        return self.__x

    @property
    def y(self):
        return self.__y
        
    def __getitem__(self, index: int):
        #todo: assertion to verify out of bounds
        return self.__x[index], self.__y[index]
    
    def __setitem__(self, index: int, x_: np.ndarray, y_: np.ndarray):
        # todo: assertion to verify out of bounds
        self.__x[index] = x_
        self.__y[index] = y_
    
    def __len__(self):
        return self.__n
    
    def __del__(self):
        del(self.__x)
        del(self.__y)
        del(self.__n)
    
    def __iter__(self):
        self.__index = 0
        return self
    
    def __next__(self):
        if (self.__index < self.__n):
            self.__index += 1
            return self[self.__index - 1]
        raise StopIteration
```

#### Generating data points

```python
# creating dataset

### setting up parameters
n = 20 # dataset size
mean, variance = 0, 0.004
interval = [0, 2*np.pi]
f = lambda x: np.cos(2*np.pi*x) + x/(2*np.pi)

def create_dataset(n, interval, f):
    x = np.array([np.random.uniform(*interval) for i in range(n)])
    y = f(x) + np.random.normal(loc = mean, scale = variance ** 0.5, size = n)
    return Dataset(x, y)

dataset = create_dataset(n, interval, f)
```

```python
for x, y in dataset:
    print(f'{x:.5f} {y:.5f}')
```

#### Polynomial Regression

```python
# feature construction

### generalized polynomial feature construction utility which returns function for
### feature construction and takes as input the degree of polynomial
polynomial_feature = lambda degree: (lambda x: np.array([x**i for i in range(degree+1)]))

### function that returns the design matrix after applying feature construction
### to the input data
def design_matrix(x, feature_function):
    return feature_function(x).T
```

```python
# regression using normal equation method

def regression(x, y):
    X = design_matrix(x, polynomial_feature(degree))
    Y = y
    w = (np.linalg.pinv(X.T @ X)) @ (X.T @ Y)
    return w

def output(weight, x):
    degree = len(weight) - 1
    X = design_matrix(x, polynomial_feature(degree))
    return (X @ weight)
```

```python
# fitting polynomial curves of different degrees
# and storing weights in a dictionary

### setting up parameters
m = [1, 2, 3, 5, 7, 10]

weights = {}

for degree in m:
    weights[degree] = regression(dataset.x, dataset.y)
```

```python
# plotting curves along with data for analysis

### function for plotting dataset
def plot_dataset(ax, dataset: Dataset, color = 'blue'):
    ax.scatter(dataset.x, dataset.y, color = color)
    
def mean_squared_error(y, target):
    return np.sum((y-target)**2) / len(y)

def plot_all_curves(dataset, weights):
    for degree in m:
        fig = plt.figure()
        ax = fig.add_subplot(111)

        # plotting dataset
        plot_dataset(ax, dataset)

        # plotting polynomial fit
        poly_x = np.linspace(*interval, num = int(100 * (interval[1] - interval[0])))
        poly_y = output(weights[degree], poly_x)

        ax.plot(poly_x, poly_y, color = 'red')

        # plotting true function
        true_x = poly_x
        true_y = f(true_x)

        ax.plot(true_x, true_y, color = 'black')

        train_error = mean_squared_error(output(weights[degree], dataset.x), dataset.y)
        test_error = mean_squared_error(output(weights[degree], true_x), true_y)
        
        plt.title(f'Degree: {degree}    Training Error: {train_error: .3f}    Testing Error: {test_error: .3f}')
        plt.xlabel('x')
        plt.ylabel('y')
        
        plt.grid('True')

        plt.show()

plot_all_curves(dataset, weights)
```

#### Observations

- As the degree of the polynomial increases, training error in general reduces as the polynomial with higher degree has more degrees of freedom because of which it is able to fit the given data better.
- Increasing the degree however, has an ill effect over the testing error. This is the result of overfitting as the datapoints given are very less.

#### Repeating with more data

```python
'''
lets sample points from the given distribution with the dataset size equal to 100 this time
'''

# setting the parameter
n = 100

# creating new dataset
dataset = create_dataset(n, interval, f)

# fitting polynomials of all the given degrees
for degree in m:
    weights[degree] = regression(dataset.x, dataset.y)

# plotting results
plot_all_curves(dataset, weights)
```

#### Observations

- With increased dataset size, now the polynomial is able to fit better to the original curve, although still not fully.
- Training error in general still decreases with increasing degree of polynomial. The reason for this being the same as stated earlier.
- Testing error however reduces with increasing degree because with increased number of datapoints, polynomial with higher degree is able to fit better now as compared to the polynomial with less degree which despite large dataset cannot fit well as it has lesser number of degrees of freedom.
